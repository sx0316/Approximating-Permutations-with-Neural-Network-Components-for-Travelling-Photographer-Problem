```python
def ve_single(state1, state2, observation):
    
    a = fnnn(state1, observation)
    b = fnnn(state2, observation)

    if not a or not b:
        return 0
    
    xs, ys = np.random.normal(a/2, np.sqrt(a/5), 5), np.random.normal(b/2, np.sqrt(b/5), 5)
    x, y = np.median(xs), np.median(ys)
    res = x*y
    if res < 0.001:
        res = 0.00

    return res

def ve_obs(obs_in, num_states):
    f = [[0 for i in range(num_states)] for i in range(num_states)]
    source, sink, middle_node, first = [], [], [], True
    obs_seq = obs_in.split(';')
    middle = False
    
    for ob in obs_seq:
        #找出所有跟这个o有关的state
        sink = []
        ob_instance = ob.split(",")
        iii = 0
        
        for inst in ob_instance:
            connected_states = find_states_connected(inst)
            if  iii== int(round(len(ob_instance) / 2)):
                middle = True
            iii += 1
   
            for s in connected_states:
                f[s][s] += 1
                if first:
                    source.append(s)
                elif middle:
                    middle_node.append(s)
                else:
                    sink.append(s)
            connected_states2 = connected_states
            for o in inst:
                wij = 0
                for i in connected_states:
                    for j in connected_states2:
                        wij = ve_single(i, j, str(o))
                        f[i][j] += wij
                        f[j][i] += wij
                            
        first, middle = False, False
    for i in range(num_states):
        f[i][i] = 0
    
    source = list(dict.fromkeys(source))
    sink = list(dict.fromkeys(sink))
    middle_node = list(set(middle_node))
    return f, source, middle_node, sink
```
